```python
import plotly.graph_objects as go
import numpy as np
np.random.seed(1)

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sz = np.random.rand(N) * 30

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode="markers",
    marker=go.scatter.Marker(
        size=sz,
        color=colors,
        opacity=0.6,
        colorscale="Viridis"
    )
))

fig.show()
```

```python
import os

if not os.path.exists("images"):
    os.mkdir("images")
```

```python
fig.write_image("images/fig1.pdf",format="pdf", engine="orca")
```

```python
import plotly.graph_objects as go

# Sample data (replace with your actual data)
data = np.random.rand(10, 10)

# Reshape data for scatter plot (optional)
x, y = np.meshgrid(np.arange(len(data[0])), np.arange(len(data)))

# Create a scatter trace with markers sized and colored based on data values
scatter = go.Scatter(
    x=x.flatten(),
    y=y.flatten(),
    mode="markers",
    marker=dict(size=4, color=data.flatten(), colorscale="viridis", opacity=0.8)
)

# Create a layout and add the scatter trace
layout = go.Layout(title="Scatter Plot Example")
fig = go.Figure(data=[scatter], layout=layout)

# Display the plot
fig.show()
```
