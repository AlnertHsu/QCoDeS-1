```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from skimage import io
from skimage.color import rgba2rgb, rgb2gray

# Load the image
img = io.imread(r'106-1Data\106-1Plot\106-1_135.png')

# Convert RGBA to RGB, then to grayscale
rgb_img = rgba2rgb(img)
gray_img = rgb2gray(rgb_img)

# Crop the image to remove axes and color bar
# You may need to adjust these values based on your image
cropped_img = gray_img[85:-105, 140:-200]

# Average the pixel values along the y-axis
avg_signal = np.mean(cropped_img, axis=0)

# Perform FFT
fft_result = np.fft.fft(avg_signal)
frequencies = np.fft.fftfreq(len(avg_signal))

# Find the peak frequency (excluding DC component)
peak_freq = frequencies[np.argmax(np.abs(fft_result[1:]))+1]

# Calculate the period
image_width = 2.0  # mT, from -1.0 to 1.0
pixels_per_mt = len(avg_signal) / image_width
period = 1 / (peak_freq * pixels_per_mt)

print(f"Estimated SQUID period: {period:.4f} mT")

# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(311)
plt.imshow(cropped_img, cmap='viridis', aspect='auto')
plt.title('Cropped Image')

plt.subplot(312)
plt.plot(avg_signal)
plt.title('Averaged Signal')

plt.subplot(313)
plt.plot(frequencies[1:], np.abs(fft_result[1:]))
plt.title('FFT Magnitude')
plt.xlabel('Frequency')

plt.tight_layout()
plt.show()
```
